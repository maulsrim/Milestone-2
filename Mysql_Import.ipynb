```python
#Convert all rows into text
y = []

for i in range(len(df)):
    x = tuple(df.iloc[i])
    y.append(x)

y
```

```python
file = open('Salary_Survey_2021.txt', 'w', encoding = 'UTF-8')

for my_tuple in y:
    file.write(str(my_tuple)+ ',' + '\n')

file.close()
```
